<h1>solucion 1 problema

```python
# importamos las columnas developer e item_id
df_dev=df_items[['developer', 'price', 'release_year']]

#borramos los valores nulos en developer
df_dev=df_dev.dropna(subset=['developer'])

df_dev=df_dev.sort_values(by='developer', ascending=True)
# Filtrar el DataFrame para excluir las filas con valores diferentes de 0 en 'price'
df_dev = df_dev[df_dev['price'] == 0]
df_dev
```

indicamos los valores de la columna developer y cuantas veces se repite

```python
df_dev['developer'].value_counts()
```

como indica muchos desarrolladores, se hace un filtro eliminando los desarrolladores que tienen 1 juego gratis

```python
valores = df_dev['developer'].value_counts() == 1

# Filtrar el DataFrame para excluir las filas con valores que se repiten solo una vez
df_dev = df_dev[~df_dev['developer'].isin(valores[valores].index)]
df_dev
```

se cuenta cuantas veces se repite cada valor de la columna developer

```python
df_dev['developer'].value_counts()
```

```python
df_dev.rename(columns={'release_year': 'Año'}, inplace=True)
```

se guarda el archivo para subir a fastapi y no use muchos recursos 

```python
df_dev.to_csv('1punto.csv', index=False)
```

se genera la solución sin fastapi

```python
def developer(desarrollador: str):
    
    # Filtrar el DataFrame para obtener solo las filas correspondientes al desarrollador especificado
    df_filtrado = df_dev[df_dev['developer'] == desarrollador]

    # Calcular la cantidad de items por año
    cantidad_items_por_anio = df_filtrado.groupby('Año').size()

    # Calcular el porcentaje de contenido gratuito por año
    porcentaje_contenido_gratuito_por_anio = (df_filtrado.groupby('Año').size())/((cantidad_items_por_anio).sum())*100

    # Crear un nuevo DataFrame con los resultados
    resultado = pd.DataFrame({'Cantidad': cantidad_items_por_anio,
                              'Contenido Free': porcentaje_contenido_gratuito_por_anio})
    
    # Formatear el porcentaje con '%' al final
    resultado['Contenido Free'] = resultado['Contenido Free'].astype(str) + '%'

    return resultado
```

```python
developer('3D Realms (Apogee Software)')
```

Se hace un merge entre los diferentes dataframe para dar solución al 2 problema

```python
# Realiza la relación entre df_user y df_item en base a 'item_id' y agrega la columna 'price' de df_item
df_datauser = pd.merge(df_user[['user_id', 'item_id']], df_items[['item_id', 'price']], on='item_id', how='left')
df_datauser = pd.merge(df_datauser, df_review[['user_id', 'item_id', 'recommend']], on=['user_id', 'item_id'], how='left')

# la columna 'user_id' se pasara a minusculas
df_datauser['user_id'] = df_datauser['user_id'].str.lower()
df_datauser
```

```python
#se revisa los valores nulos en la columna
df_datauser['recommend'].isnull().sum()
```

```python
# Reemplaza los valores nulos en la columna 'recommend' por False
df_datauser['recommend'].fillna(False, inplace=True)
```

se borra la columna item_id para liberar espacio

```python
df_datauser = df_datauser.drop(columns=['item_id'])
```

se genera un parquet para utilizar con fastapi

```python

df_datauser.to_parquet('2punto.parquet', index=False)
```

```python
def userdata(user_id: str):
    user_id = user_id.lower()
    comprobar = list(set(df_datauser['user_id']))
    if user_id in comprobar:
           
         # Filtrar el DataFrame por el usuario especificado
        user_data = df_datauser[df_datauser['user_id'] == user_id]
        
        # Contar las repeticiones del usuario
        contador = user_data.shape[0]
        
        # Sumar la columna 'price' para las filas correspondientes al usuario
        suma_precios = user_data['price'].sum()
        # Filtrar las filas donde la columna 'recommend' es True y luego sumar esos valores para dividirlos entre el contador
        promedio = (user_data[user_data['recommend'] == True]['recommend'].sum())*100/contador


        print(f"El usuario {user_id} tiene {contador} items.")
        print(f"La suma dinero gastado para este usuario es: {suma_precios}")
        print(f"El porcentaje de recomendaciones para este usuario es: {promedio:.2f} %")
    
        return 
    else:
        print('Ingrese nuevamente el usuario')
        return 
```

<h1> Solución problema 3

def UserForGenre( genero : str ): Debe devolver el usuario que acumula más horas jugadas para el género dado y una lista de la acumulación de horas jugadas por año de lanzamiento analizado bajo la columna PlayTimeForever.

```python
# Realiza la relación entre df_user y df_item en base a 'item_id' 
df_genre = pd.merge(df_user, df_items, on='item_id', how='right')
df_genre
```

Se elimina las filas que contienen nulos en la columna user_id y se convierte la columna playtime_forever a int32 para reducir espacio

```python
df_genre=df_genre.dropna(subset=['user_id'])

df_genre['playtime_forever'] = df_genre['playtime_forever'].astype('int32')

df_genre
```

Se eliminan las columnas 'publisher', 'title', 'price', 'developer', 'item_id' que no se van a usar

```python
df_genre = df_genre.drop(columns=['publisher', 'title', 'price', 'developer', 'item_id'])
df_genre
```

se borra las filas que tienen valores 0 en las columnas playtime_forever, ya que no se las usara 

```python
df_genre = df_genre.loc[~((df_genre['playtime_forever'] == 0) )]
df_genre
```

```python
# Reemplaza True por su equivalente en la columna "playtime_forever" para las de los generos para su posterior analisis
df_genre[df_genre.columns[3:]] = df_genre[df_genre.columns[3:]].multiply(df_genre['playtime_forever'], axis='index')

df_genre
```

```python
#elimina la columna playtime_forever
df_genre = df_genre.drop(columns=['playtime_forever'])
```

```python
df_genre.to_parquet('3punto.parquet', index=False)
```

```python
def UserForGenre( genero : str ):


    # Encontrar el usuario con más horas jugadas para el género dado
    usuario_mas_horas = df_genre.groupby('user_id')[genero].sum().idxmax()
    

    # Filtrar el DataFrame para el usuario con más horas jugadas
    df_usuario_mas_horas = df_genre[df_genre['user_id'] == usuario_mas_horas]
    
    # Agrupar por año y sumar las horas jugadas
    sumas_por_anio = df_usuario_mas_horas.groupby('release_year')[genero].sum()

    
    # Crear la estructura de retorno
    retorno = {
        "Usuario con más horas jugadas para Género": usuario_mas_horas,
        "Horas jugadas": sumas_por_anio.to_dict
    }

    return {
        "Usuario con más horas jugadas del genero": usuario_mas_horas,
        "Horas jugadas": sumas_por_anio        
    }   
```

```python
df_review
```

en reviews.ipynb se reviso que quienes tengan un 2 en analysis de sentimientos son positivos

```python
df_best = df_review.loc[(df_review['recommend'] == True) & (df_review['sentiment_analysis'] == 2)]
df_best=df_best.drop(columns=['user_id'])
df_best
```

```python
df_items
```

se hacen un merge entre los df_best, df_items par extraer el nombre de los desarrolladores

```python
df_b_dev = pd.merge(df_best, df_items[['item_id', 'developer']], on='item_id', how='left')
df_b_dev
```

```python
df_b_dev.to_csv('4punto.csv', index=False)
```

la función encuentra y muestra el top 3 de desarrolladores con juegos MÁS recomendados por usuarios para el año dado. (reviews.recommend = True y comentarios positivos)

```python
def best_developer_year( año : int ):
    
    df_filt = df_b_dev.loc[(df_b_dev['year'] == año)]
    
    df_agrupado = df_filt.groupby('developer').size().reset_index(name='repeticiones')
    
    df_top3 = df_agrupado.sort_values('repeticiones', ascending=False).head(3).to_dict
    
    
    return df_top3
  
```
